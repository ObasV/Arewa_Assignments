```python
# Plot the first five cubic numbers

import matplotlib.pyplot as plt

numbers = list(range(6))
cubed_numbers = list(map(lambda x: x**3, numbers))

plt.style.use('seaborn-v0_8-bright')
fig, ax = plt.subplots()

ax.scatter(numbers, cubed_numbers, s = 100,  c= cubed_numbers , cmap=plt.cm.Blues)

# Set chart title and label axes.
ax.set_title("Cubed Numbers 0 - 5", fontsize=24)
ax.set_xlabel("Numbers", fontsize=14)
ax.set_ylabel("Cube of Numbers", fontsize=14)

# Set size of tick labels.
ax.tick_params(labelsize=14)


```

```python
# Plot the first 5000 cubic numbers

numbers = list(range(0,5000))
cubed_numbers = [x**3 for x in numbers]

plt.style.use('dark_background')
fig, ax = plt.subplots()

ax.scatter(numbers, cubed_numbers, s = 100,  c= cubed_numbers, cmap=plt.cm.Blues,)

# Set chart title and label axes.
ax.set_title("Cubed Numbers 0 - 5000", fontsize=24)
ax.set_xlabel("Numbers", fontsize=14)
ax.set_ylabel("Cube of Numbers", fontsize=14)

# Set size of tick labels.
ax.tick_params(labelsize=14)
```
